```python
from aixplain.factories import IndexFactory
from dotenv import find_dotenv, load_dotenv
from aixplain.modules.model.record import Record
import os
import csv

load_dotenv(find_dotenv())
API_KEY = os.getenv("AIXPLAIN_API_KEY")
```

```python

csv_path = "../data/gdpr_text.csv"
Records = []
with open(csv_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Records.append(
            Record(
                value=row["gdpr_text"],
                value_type="text",
                id=row["chapter"],
                attributes={
                    "chapter_title": row["chapter_title"],
                    "article_title": row["article_title"],
                    "sub_article": row["sub_article"],
                    "href": row["href"],
                }
            )
        )
print(Records)
```

```python
index = IndexFactory.create(
    name="GDPR Policy Index 2",
    description="Vector index for GDPR documents from CSV",
)
print(index)
```

```python
# Agent to retrive the data

from aixplain.factories import AgentFactory

agent = AgentFactory.create(
    name="Knowledge Assistant",
    description="Answers based on the provided index",
    tools=[
        AgentFactory.create_model_tool(index.id)
    ]
)

response = agent.run(" processing of personal data and rules relating to the free movement of personal data.")
print("index",index.id)
print(response.data.output)
```
